```python
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
 df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
```

```python
df.info()
```

```python
df['Age'] = 2026 - df['Year_Birth']
```

```python
df[['Year_Birth','Age']].head()
```

```python
df['Total_Spending'] = (
    df['MntWines']
    + df['MntFruits']
    + df['MntMeatProducts']
    + df['MntFishProducts']
    + df['MntSweetProducts']
    + df['MntGoldProds']
)
```

```python
df[['Total_Spending']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=20)

plt.title("Age Distribution of Customers")

plt.show()
```

```python
df['Total_Spending'] = (
    df['MntWines'] +
    df['MntFruits'] +
    df['MntMeatProducts'] +
    df['MntFishProducts'] +
    df['MntSweetProducts'] +
    df['MntGoldProds']
)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Income'], bins=20)

plt.title("Income Distribution")

plt.show()
```

```python
campaigns = [
    'AcceptedCmp1',
    'AcceptedCmp2',
    'AcceptedCmp3',
    'AcceptedCmp4',
    'AcceptedCmp5'
]

df[campaigns].sum().plot(kind='bar')

plt.title("Campaign Acceptance")

plt.show()
```

```python
products = [
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]

df[products].sum().plot(kind='bar')

plt.title("Product Spending by Category")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Income'], df['Total_Spending'])

plt.xlabel("Income")
plt.ylabel("Total Spending")
plt.title("Income vs Total Spending")

plt.show()
```

```python
df['Response'].value_counts().plot(kind='bar')

plt.title("Campaign Response")

plt.show()
```
